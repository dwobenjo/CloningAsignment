```python
import pandas as pd
import seaborn as sns

# Load Titanic dataset
df = sns.load_dataset('titanic')
df.head()
```

```python
df.columns
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Pie chart of survival

# Count survival values
survival_counts = df['survived'].value_counts()

# Labels for the chart
labels = ['Did not survive', 'Survived']

# Set Seaborn style
sns.set_theme(style="whitegrid")

# Create a pie chart
plt.figure(figsize=(6, 6))
plt.pie(survival_counts, labels=labels, autopct='%1.1f%%', colors=sns.color_palette("pastel"), startangle=90)
plt.title('Survival Distribution on Titanic')
plt.show()
```

```python
# barchart 

# Set the theme
sns.set_theme(style="whitegrid")

# Create a bar chart for passenger class and sex
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x="pclass", hue="sex", palette="pastel")

# Set labels
plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.title("Passenger Class Distribution by Sex on Titanic")

# Show plot
plt.show()
```

```python
# histogram

# Set the theme
sns.set_theme(style="whitegrid")

# Create a histogram for age distribution
plt.figure(figsize=(8, 5))
sns.histplot(df['age'], bins=30, kde=True, color="blue")

# Set labels
plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Age Distribution of Titanic Passengers")

# Show plot
plt.show()
```
